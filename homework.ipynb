```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузка данных
kc_house_data = pd.read_csv('kc-house-data.csv')

# Просмотр первых строк данных
kc_house_data_head = kc_house_data.head()

kc_house_data_head
```

```python
# Распределение домов по наличию вида на набережную
waterfront_counts = kc_house_data['waterfront'].value_counts()

# Создание DataFrame для удобства работы с Seaborn
waterfront_df = waterfront_counts.reset_index()
waterfront_df.columns = ['waterfront', 'count']

# Построение графика
plt.figure(figsize=(8, 6))
sns.barplot(data=waterfront_df, x='waterfront', y='count', hue='waterfront', dodge=False, palette='Blues')
plt.title('Распределение домов по наличию вида на набережную')
plt.xlabel('Наличие вида на набережную (1 - да, 0 - нет)')
plt.ylabel('Количество домов')
plt.xticks([0, 1], ['Нет', 'Да'])
plt.legend(title='Вид на набережную', loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Распределение этажей домов
floors_counts = kc_house_data['floors'].value_counts().sort_index()

# Создание DataFrame
floors_df = floors_counts.reset_index()
floors_df.columns = ['floors', 'count']

# Построение графика
plt.figure(figsize=(10, 6))
sns.barplot(data=floors_df, x='floors', y='count', hue='floors', dodge=False, palette='viridis')
plt.title('Распределение этажей домов в King County')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.legend(title='Этажи', loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Распределение состояния домов
condition_counts = kc_house_data['condition'].value_counts().sort_index()

# Создание DataFrame
condition_df = condition_counts.reset_index()
condition_df.columns = ['condition', 'count']

# Построение графика
plt.figure(figsize=(10, 6))
sns.barplot(data=condition_df, x='condition', y='count', hue='condition', dodge=False, palette='magma')
plt.title('Распределение состояния домов в King County')
plt.xlabel('Состояние дома (1 - очень плохое, 5 - отличное)')
plt.ylabel('Количество домов')
plt.legend(title='Состояние', loc='upper right')
plt.grid(True)
plt.show()
```
