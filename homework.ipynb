```python
import pandas as pd

# Загрузка данных
kc_house_data = pd.read_csv('kc-house-data.csv')

# Просмотр первых строк данных
kc_house_data_head = kc_house_data.head()

kc_house_data_head

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Распределение цен на недвижимость
plt.figure(figsize=(10, 6))
sns.histplot(kc_house_data['price'], bins=50, kde=True)
plt.title('Распределение цен на недвижимость в King County')
plt.xlabel('Цена (в долларах)')
plt.ylabel('Количество объектов')
plt.grid(True)
plt.show()
```

```python
# Распределение площадей жилой недвижимости
plt.figure(figsize=(10, 6))
sns.histplot(kc_house_data['sqft_living'], bins=50, kde=True)
plt.title('Распределение площадей жилой недвижимости в King County')
plt.xlabel('Площадь жилой недвижимости (кв. футы)')
plt.ylabel('Количество объектов')
plt.grid(True)
plt.show()
```

```python
# Распределение годов постройки
plt.figure(figsize=(10, 6))
sns.histplot(kc_house_data['yr_built'], bins=50, kde=True)
plt.title('Распределение года постройки недвижимости в King County')
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов')
plt.grid(True)
plt.show()
```
