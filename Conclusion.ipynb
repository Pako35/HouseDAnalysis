```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузка данных
kc_house_data = pd.read_csv('kc-house-data.csv')

# Просмотр первых строк данных
kc_house_data_head = kc_house_data.head()

kc_house_data_head
```

```python
# Построение scatter-плот для площади жилой недвижимости и стоимости
plt.figure(figsize=(10, 6))
sns.scatterplot(data=kc_house_data, x='sqft_living', y='price', hue='price', palette='viridis', alpha=0.6)
plt.title('Зависимость стоимости от площади жилой недвижимости')
plt.xlabel('Площадь жилой недвижимости (кв. футы)')
plt.ylabel('Стоимость (в долларах)')
plt.grid(True)
plt.show()
```

```python
# Построение box-плота для вида на набережную и стоимости
plt.figure(figsize=(8, 6))
sns.boxplot(data=kc_house_data, x='waterfront', y='price', palette='Blues')
plt.title('Влияние вида на набережную на стоимость недвижимости')
plt.xlabel('Вид на набережную (0 - нет, 1 - да)')
plt.ylabel('Стоимость (в долларах)')
plt.grid(True)
plt.show()
```

```python
# Построение line-плота для года постройки и средней стоимости
avg_price_by_year_built = kc_house_data.groupby('yr_built')['price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_price_by_year_built, x='yr_built', y='price', marker='o')
plt.title('Зависимость средней стоимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Средняя стоимость (в долларах)')
plt.grid(True)
plt.show()
```

```python
# Построение box-плота для количества этажей и стоимости
plt.figure(figsize=(10, 6))
sns.boxplot(data=kc_house_data, x='floors', y='price', palette='viridis')
plt.title('Влияние количества этажей на стоимость недвижимости')
plt.xlabel('Количество этажей')
plt.ylabel('Стоимость (в долларах)')
plt.grid(True)
plt.show()
```

```python
# Построение box-плота для состояния дома и стоимости
plt.figure(figsize=(10, 6))
sns.boxplot(data=kc_house_data, x='condition', y='price', palette='magma')
plt.title('Влияние состояния дома на стоимость недвижимости')
plt.xlabel('Состояние дома (1 - очень плохое, 5 - отличное)')
plt.ylabel('Стоимость (в долларах)')
plt.grid(True)
plt.show()
```
